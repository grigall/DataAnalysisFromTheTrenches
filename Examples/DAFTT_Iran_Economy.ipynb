```python
import os
import pandas as pd

PATH = os.getcwd()
FILE = 'Iran_Economy_2018-2024.csv'
FULL_PATH = os.path.join(PATH, FILE)
data = pd.read_csv(FULL_PATH, sep='\t')
data
```

```python
import math

def round_up(n, decimals=0):
    multiplier = 10 ** decimals
    return math.ceil(n * multiplier) / multiplier

def round_down(n, decimals=0):
    multiplier = 10 ** decimals
    return math.floor(n * multiplier) / multiplier

max_val = int(round_up(round(max(data.max().tolist()[1:])),-1))
min_val = int(round_down(round(min(data.min().tolist()[1:])),-1))

print(max_val)
print(min_val)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(16,9))

plt.plot(data.iloc[:, 1:].transpose())
plt.xlabel('Years')
plt.ylabel('% of Change')
plt.xticks(ticks=np.arange(6), labels=data.columns.tolist()[1:])
plt.yticks(ticks=np.arange(min_val, max_val, 10))
plt.legend(loc='best', labels=data.loc[:, 'Category'])
plt.show()
```
